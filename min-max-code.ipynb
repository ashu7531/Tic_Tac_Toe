```python
def list_to_string(lst):
    string = ''.join(lst)
    return string
```

```python
def player_turn(state):
    num = state.count('-')
    return 'O' if num % 2 == 0 else 'X'
```

```python
def is_terminal(state):
    win_conditions = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Rows
        (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Columns
        (0, 4, 8), (2, 4, 6)              # Diagonals
    ]

    for condition in win_conditions:
        if state[condition[0]] == state[condition[1]] == state[condition[2]] == 'X':
            return True, 1
        elif state[condition[0]] == state[condition[1]] == state[condition[2]] == 'O':
            return True, -1

    if state.count('-') == 0:
        return True, 0
    else:
        return False, 0
```

```python
tree = {}
queue = [['-'] * 9]
terminal = []
child_parent = {}
Scores = {}

def create_tree(state=['-'] * 9):
    global tree, queue
    queue.pop(0)
    string_of_state = list_to_string(state)
    tree[string_of_state] = []
    turn = player_turn(state)

    if not is_terminal(state)[0]:
        for i in range(9):
            state_copy = state.copy()
            if state[i] == '-':
                state_copy[i] = turn
                state_copy_string = list_to_string(state_copy)
                tree[string_of_state].append(state_copy_string)
                queue.append(state_copy)

                if state_copy_string not in child_parent.keys():
                    child_parent[state_copy_string] = []
                score = is_terminal(state_copy_string)[1]
                Scores[state_copy_string] = [score, turn]
                if string_of_state not in child_parent[state_copy_string]:
                    child_parent[state_copy_string].append(string_of_state)

                create_tree(state_copy)
    else:
        if string_of_state not in terminal:
            terminal.append(string_of_state)

create_tree()
child_parent['---------'] = [None]
Scores['---------'] = [0, 'O']
```

```python
def update_helper(parents):
    while parents[0] is not None:
        for j in parents:
            score_list = [Scores[k][0] for k in tree[j]]
            Scores[j][0] = min(score_list) if Scores[j][1] == 'X' else max(score_list)
            parents = child_parent[j]
            update_helper(parents)

def update_values():
    for i in terminal:
        parents = child_parent[i]
        update_helper(parents)

update_values()
```

```python
def position_for_move(state, action):
    for i in range(9):
        if state[i] == '-' and action[i] != '-':
            return i

def best_move(state):
    state_str = list_to_string(state)
    children = tree[state_str]
    turn = Scores[state_str][1]

    score_mapping = {Scores[child][0]: child for child in children}
    
    best_score = min(score_mapping.keys())
    best_move_str = score_mapping[best_score]

    return position_for_move(state_str, best_move_str)
```

```python
def user_move_index():
    row = int(input("Enter the row [1, 2, 3]: "))
    column = int(input("Enter the column [1, 2, 3]: "))
    
    if row == 1:
        index = column - 1
    elif row == 2:
        index = 2 + column
    else:
        index = 5 + column
    
    return index
```

```python
def display(state):
    print('-------------------')
    print('|', state[0], '|', state[1], '|', state[2], '|')
    print('-------------------')
    print('|', state[3], '|', state[4], '|', state[5], '|')
    print('-------------------')
    print('|', state[6], '|', state[7], '|', state[8], '|')
    print('-------------------')
```

```python
def play_game():
    state = ['-'] * 9
    display(state)
    player = 0
    
    while not is_terminal(state)[0]:
        if player % 2 == 0:
            index = user_move_index()
            state[index] = 'X'
        else:
            index = best_move(state)
            state[index] = 'O'
            
        display(state)
        player += 1

        terminal = is_terminal(state)
        if terminal[1] == -1:
            return 'Computer won'
        elif terminal[1] == 1:
            return 'You won'
        elif terminal[0] and terminal[1] == 0:
            return 'Match draw'
```
